```python
import pandas as pd

# Read the CSV file
df = pd.read_csv('wetter.csv')

# Display the first few rows of the dataframe
df.head()

temperatur_series = df['Temperatur']
import matplotlib.pyplot as plt

# Plot the temperature series
plt.figure(figsize=(12, 6))
plt.plot(temperatur_series, label='Temperature')
plt.xlabel('Time')
plt.ylabel('Temperature (°C)')
plt.title('Temperature Over Time')
plt.legend()
plt.show()

# Calculate the mean temperature
df['Temperatur'].mean()
```

```python
# Add Month as string
df['Month'] = pd.to_datetime(df['Datum']).dt.strftime('%B')
df

# Calculate the mean temperature for July
mean_july_temperature = df[df['Month'] == 'July']['Temperatur'].mean()
print(mean_july_temperature)

# Calculate the mean temperature for May
mean_may_temperature = df[df['Month'] == 'May']['Temperatur'].mean()
print(mean_may_temperature)
```

```python
from scipy.stats import f_oneway

# Calculate the mean temperature for July
july_temperature = df[df['Month'] == 'July']['Temperatur']

# Calculate the mean temperature for May
may_temperature = df[df['Month'] == 'May']['Temperatur']

# Calculate the mean temperature for July
# Perform ANOVA
f_statistic, p_value = f_oneway(july_temperature, may_temperature)

print(f"F-statistic: {f_statistic}")
print(f"P-value: {p_value}")

plot = df.boxplot('Temperatur', by='Month', figsize=(12, 6))

```
