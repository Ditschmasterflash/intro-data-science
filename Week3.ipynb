```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('umsatzdaten_gekuerzt.csv')

df = df.groupby('Datum').agg({'Umsatz': 'sum'}).reset_index()

df_wetter = pd.read_csv('wetter.csv')

merged_df = pd.merge(df, df_wetter, on='Datum', how='inner')

print(merged_df.head())
print(merged_df['Windgeschwindigkeit'].describe())

missing_values = merged_df['Windgeschwindigkeit'].isnull().sum()
print(f"Missing values in Windgeschwindigkeit: {missing_values}")

datum_counts = merged_df['Datum'].value_counts()
print(datum_counts)

plt.scatter(merged_df['Datum'], merged_df['Windgeschwindigkeit'])
plt.xlabel('Datum')
plt.ylabel('Windgeschwindigkeit')
plt.title('Scatter Plot of Windgeschwindigkeit')
plt.show()

plt.scatter(merged_df['Windgeschwindigkeit'], merged_df['Umsatz'])
plt.xlabel('Windgeschwindigkeit')
plt.ylabel('Umsatz')
plt.title('Scatter Plot of Umsatz vs Windgeschwindigkeit')
plt.show()

```
